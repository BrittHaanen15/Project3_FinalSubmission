```python
from tensorflow import keras

class CustomModel(keras.Model):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.loss_tracker = keras.metrics.Mean(name="loss")
        self.mae_metric = keras.metrics.MeanAbsoluteError(name="mae")

    def train_step(self, data):
        x, y, start_date, end_date = data.drop(columns=['POINT_BALANCE', 'FROM_DATE', 'TO_DATE', 'time']), data['POINT_BALANCE'], data['FROM_DATE'], data['TO_DATE']

        with tf.GradientTape() as tape:
            y_pred = self(x, training=True)  # Forward pass
            # Compute our own loss
            loss = keras.losses.mean_squared_error(y, y_pred)

        # Compute gradients
        trainable_vars = self.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        # Update weights
        self.optimizer.apply_gradients(zip(gradients, trainable_vars))

        # Compute our own metrics
        self.loss_tracker.update_state(loss)
        self.mae_metric.update_state(y, y_pred)
        return {"loss": self.loss_tracker.result(), "mae": self.mae_metric.result()}

    @property
    def metrics(self):
        # We list our `Metric` objects here so that `reset_states()` can be
        # called automatically at the start of each epoch
        # or at the start of `evaluate()`.
        # If you don't implement this property, you have to call
        # `reset_states()` yourself at the time of your choosing.
        return [self.loss_tracker, self.mae_metric]


# Construct an instance of CustomModel
inputs = keras.Input(shape=(32,))
outputs = keras.layers.Dense(1)(inputs)
model = CustomModel(inputs, outputs)

# We don't pass a loss or metrics here.
model.compile(optimizer="adam")

# Just use `fit` as usual -- you can use callbacks, etc.
x = np.random.random((1000, 32))
y = np.random.random((1000, 1))
model.fit(x, y, epochs=5)
```

```python
X = df_copy.drop(['POINT_BALANCE', 'FROM_DATE', 'TO_DATE', 'time'], axis=1).values
y = df_copy['POINT_BALANCE'].values.reshape(-1, 1)

print('Shape of X and y:', X.shape, y.shape)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)
```

```python
class deep_learning_model():
    
    def __init__(self, max_dense_layers=3):
        '''
        Initialize the deep learning model with the specified parameters.
        
        Args:
        - max_dense_layers (int): Maximum number of Dense layers.
        '''
        self.max_dense_layers = max_dense_layers

    def get_model_hyp(self, hp):
        '''
        Construct and compile the deep learning model with hyperparameter tuning.
        
        Args:
        - hp (keras_tuner.HyperParameters): Hyperparameters for tuning the model architecture.
        
        Returns:
        - model (keras.Model): Constructed and compiled CNN-LSTM model.
        '''
        # Initialize the model
        model = Sequential()
        
        # Add Dense layers with hyperparameter tuning
        for dense_layer in range(hp.Int('num_dense_layers', min_value=1, max_value=self.max_dense_layers)):
            model.add(Dense(hp.Int(f'dense_units_{dense_layer}',
                                   min_value=32,
                                   max_value=512,
                                   step=32),
                                   activation=hp.Choice(f'dense_activation_{dense_layer}',
                                                       values=['relu', 'tanh', 'sigmoid', 'linear'])))
        
        model.add(Flatten())
        # Add Output layer
        model.add(Dense(1))
        
        # Compile the model with tuned learning rate
        model.compile(optimizer=Adam(
            learning_rate=hp.Float('lr', min_value=1e-4, max_value=1e-2, step=1e-1)),
            loss='mae',
            metrics=['mae'])
        
        return model
```

```python
model = deep_learning_model(max_dense_layers=5)
```

```python
tuner = kt.RandomSearch(
            model.get_model_hyp,
            max_trials=20,
            objective='val_loss',
            overwrite = True,
            directory = './Models',
            project_name='deep_learning_model'
            )
tuner.search(X_train, y_train, epochs=500,
            validation_split=0.2,
            use_multiprocessing = True,
            workers=-1, verbose = 1,
            callbacks=[EarlyStopping(monitor='val_loss', patience=10)]
            )
```

```python
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
# Inverse transform the scaled data
y_train_inv = scaler.inverse_transform(y_train)
y_pred_train_inv = scaler.inverse_transform(best_model.predict(X_train))
y_test_inv = scaler.inverse_transform(y_test)
y_pred_test_inv = scaler.inverse_transform(best_model.predict(X_test))

# Calculate metrics
train_corr = np.corrcoef(y_train_inv.flatten(), y_pred_train_inv.flatten())[0, 1]
test_corr = np.corrcoef(y_test_inv.flatten(), y_pred_test_inv.flatten())[0, 1]
train_mae = mean_absolute_error(y_train_inv, y_pred_train_inv)
test_mae = mean_absolute_error(y_test_inv, y_pred_test_inv)
train_r2 = r2_score(y_train_inv, y_pred_train_inv)
test_r2 = r2_score(y_test_inv, y_pred_test_inv)

# Print metrics
print(f"Training Correlation: {train_corr:.4f}")
print(f"Testing Correlation: {test_corr:.4f}")
print(f"Training MAE: {train_mae:.4f}")
print(f"Testing MAE: {test_mae:.4f}")
print(f"Training R Score: {train_r2:.4f}")
print(f"Testing R Score: {test_r2:.4f}")
```

```python
# Define the model
xgb_model = XGBRegressor()

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.0001, 0.001, 0.01, 0.1, 0.2],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
}

# Define the grid search
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error', verbose=1, n_jobs=-1)

# Fit the grid search
grid_search.fit(X_train, np.ravel(y_train))

# Get the best model
best_xgb_model = grid_search.best_estimator_
clear_output()
# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)

# Predict and evaluate the model
y_pred_train_xgb = best_xgb_model.predict(X_train)
y_pred_test_xgb = best_xgb_model.predict(X_test)

# Inverse transform the scaled data
y_pred_train_xgb_inv = scaler.inverse_transform(y_pred_train_xgb.reshape(-1, 1))
y_pred_test_xgb_inv = scaler.inverse_transform(y_pred_test_xgb.reshape(-1, 1))

# Calculate metrics
train_corr_xgb = np.corrcoef(y_train_inv.flatten(), y_pred_train_xgb_inv.flatten())[0, 1]
test_corr_xgb = np.corrcoef(y_test_inv.flatten(), y_pred_test_xgb_inv.flatten())[0, 1]
train_mae_xgb = mean_absolute_error(y_train_inv, y_pred_train_xgb_inv)
test_mae_xgb = mean_absolute_error(y_test_inv, y_pred_test_xgb_inv)
train_r2_xgb = r2_score(y_train_inv, y_pred_train_xgb_inv)
test_r2_xgb = r2_score(y_test_inv, y_pred_test_xgb_inv)

# Print metrics
print(f"Training Correlation (XGBoost): {train_corr_xgb:.4f}")
print(f"Testing Correlation (XGBoost): {test_corr_xgb:.4f}")
print(f"Training MAE (XGBoost): {train_mae_xgb:.4f}")
print(f"Testing MAE (XGBoost): {test_mae_xgb:.4f}")
print(f"Training R Score (XGBoost): {train_r2_xgb:.4f}")
print(f"Testing R Score (XGBoost): {test_r2_xgb:.4f}")
```
